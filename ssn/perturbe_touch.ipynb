```python
import numpy as np
import cv2

import matplotlib.pyplot as plt
```

```python
def show(img):
    plt.figure(figsize=(15,10))
    plt.imshow(img)
    plt.show()
```

```python
test_touch = '/home/ysheng/Dataset/human_ds_render/touch/combine_female_genesis3_basic_toulouse_combinenomadtco_Fashion_Model_Pose_08/pitch_30_rot_45_fov_86_touch.png'
test_touch_img = plt.imread(test_touch)
show(test_touch_img)
```

```python
# def perturb(img):
#     # net input perturbation

kernel = np.ones((2,2))

erosion = cv2.erode(test_touch_img, kernel, iterations = 1)
dilation = cv2.dilate(test_touch_img, np.ones((5,5)),iterations = 1)
opening = cv2.morphologyEx(test_touch_img, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(test_touch_img, cv2.MORPH_CLOSE, kernel)
gradient = cv2.morphologyEx(test_touch_img, cv2.MORPH_GRADIENT, kernel)
tophat = cv2.morphologyEx(test_touch_img, cv2.MORPH_TOPHAT, kernel)
blackhat = cv2.morphologyEx(test_touch_img, cv2.MORPH_BLACKHAT, kernel)

norm_grad = gradient
cv2.normalize(gradient, norm_grad, 0.0, 1.0, cv2.NORM_MINMAX)

norm_tophat = tophat
cv2.normalize(tophat, norm_tophat, 0.0, 1.0, cv2.NORM_MINMAX)

norm_blackhat = blackhat
cv2.normalize(blackhat, norm_blackhat, 0.0, 1.0, cv2.NORM_MINMAX)

plt.figure(figsize=(20,15))
plt.subplot(3,3,1)
plt.imshow(erosion)

plt.subplot(3,3,2)
plt.imshow(dilation)

plt.subplot(3,3,3)
plt.imshow(opening)

plt.subplot(3,3,4)
plt.imshow(closing)

plt.subplot(3,3,5)
plt.imshow(norm_grad)

plt.subplot(3,3,6)
plt.imshow(norm_tophat)

plt.subplot(3,3,7)
plt.imshow(norm_blackhat)
```
